```python
import tkinter as tk
from tkinter import ttk
import mysql.connector

# MySQL database credentials
db_config = {
    'user': 'root',
    'password': 'Anejs@mi624',
    'host': 'localhost',
    'database': 'formula1',  # Change to your F1 dataset database name
}

# Function to fetch and display driver standings in a table
def show_driver_standings_table(root):
    # Connect to the database
    connection = mysql.connector.connect(**db_config)
    cursor = connection.cursor()

    # Execute query to fetch driver standings
    cursor.execute("SELECT d.driverid, d.forename AS driver_name, SUM(ds.points) AS total_points FROM drivers d INNER JOIN driver_standings ds ON d.driverid = ds.driverid GROUP BY d.driverid, d.forename ORDER BY total_points DESC;")
    driver_standings = cursor.fetchall()

    # Close cursor and connection
    cursor.close()
    connection.close()

    # Create a new window to display results
    driver_window = tk.Toplevel(root)
    driver_window.title("Driver Standings")

    # Create a treeview widget
    tree = ttk.Treeview(driver_window)
    tree["columns"] = ("Driver ID", "Driver Name", "Total Points")
    tree.heading("#0", text="Rank")
    tree.heading("Driver ID", text="Driver ID")
    tree.heading("Driver Name", text="Driver Name")
    tree.heading("Total Points", text="Total Points")

    # Insert data into the treeview
    for index, row in enumerate(driver_standings, start=1):
        tree.insert("", "end", text=index, values=row)

    # Pack the treeview
    tree.pack(expand=True, fill="both")

# Function to fetch and display constructor standings in a table
def show_constructor_standings_table(root):
    # Connect to the database
    connection = mysql.connector.connect(**db_config)
    cursor = connection.cursor()

    # Execute query to fetch constructor standings
    cursor.execute("SELECT c.constructorid, c.name AS constructor_name, SUM(cs.points) AS total_points FROM constructor c INNER JOIN constructor_standings cs ON c.constructorid = cs.constructorid GROUP BY c.constructorid, c.name ORDER BY total_points DESC;")
    constructor_standings = cursor.fetchall()

    # Close cursor and connection
    cursor.close()
    connection.close()

    # Create a new window to display results
    constructor_window = tk.Toplevel(root)
    constructor_window.title("Constructor Standings")

    # Create a treeview widget
    tree = ttk.Treeview(constructor_window)
    tree["columns"] = ("Constructor ID", "Constructor Name", "Total Points")
    tree.heading("#0", text="Rank")
    tree.heading("Constructor ID", text="Constructor ID")
    tree.heading("Constructor Name", text="Constructor Name")
    tree.heading("Total Points", text="Total Points")

    # Insert data into the treeview
    for index, row in enumerate(constructor_standings, start=1):
        tree.insert("", "end", text=index, values=row)

    # Pack the treeview
    tree.pack(expand=True, fill="both")

# Function to fetch and display circuits in a table
def show_circuits_table(root):
    # Connect to the database
    connection = mysql.connector.connect(**db_config)
    cursor = connection.cursor()

    # Execute query to fetch circuits
    cursor.execute("SELECT circuitid, circuitref, name, location, country FROM circuits;")
    circuits = cursor.fetchall()

    # Close cursor and connection
    cursor.close()
    connection.close()

    # Create a new window to display results
    circuits_window = tk.Toplevel(root)
    circuits_window.title("Circuits")

    # Create a treeview widget
    tree = ttk.Treeview(circuits_window)
    tree["columns"] = ("Circuit ID", "Circuit Ref", "Name", "Location", "Country")
    tree.heading("#0", text="Rank")
    tree.heading("Circuit ID", text="Circuit ID")
    tree.heading("Circuit Ref", text="Circuit Ref")
    tree.heading("Name", text="Name")
    tree.heading("Location", text="Location")
    tree.heading("Country", text="Country")

    # Insert data into the treeview
    for index, row in enumerate(circuits, start=1):
        tree.insert("", "end", text=index, values=row)

    # Pack the treeview
    tree.pack(expand=True, fill="both")

# Main function to create GUI
def main():
    root = tk.Tk()
    root.title("F1 Dataset Viewer")

    # Dropdown menu to select query type
    query_var = tk.StringVar()
    query_dropdown = ttk.Combobox(root, textvariable=query_var, values=["Driver Standings", "Constructor Standings", "Circuits"])
    query_dropdown.pack(pady=10)

    # Button to execute selected query
    execute_button = tk.Button(root, text="Execute", command=lambda: execute_query(root, query_var))
    execute_button.pack(pady=5)

    root.mainloop()

# Function to execute selected query
def execute_query(root, query_var):
    selected_query = query_var.get()

    if selected_query == "Driver Standings":
        show_driver_standings_table(root)
    elif selected_query == "Constructor Standings":
        show_constructor_standings_table(root)
    elif selected_query == "Circuits":
        show_circuits_table(root)

if __name__ == "__main__":
    main()
```
